### Categorical Features 

```python
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype =='O']
```

```python
for feature in feature_nan:
    print("Feature is {} and values missing percentage {}% ".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
def repl_categoricalna(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna("Missing")
    return data
dataset=repl_categoricalna(dataset,feature_nan)
```

```python
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
feature_contnan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype !='O']
```

```python
for feature in feature_contnan:
    print("Missing feature {} and Values are {}% ",feature,np.round(dataset[feature].isnull().mean(),4))
```

```python
#Replacng missing values of numerical feature with median

for feature in feature_contnan:
    median_value=dataset[feature].median()
    #create a new feature to replace it with 1 an d0
    dataset[feature+"na"]=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
```
